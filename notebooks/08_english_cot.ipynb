```python
from vllm import LLM, SamplingParams
model_name = "allenai/OLMo-2-1124-7B-Instruct"
llm = LLM(
    model=model_name,
    dtype="float16",
    gpu_memory_utilization=0.85
)
sampling_params = SamplingParams(
    temperature=0.0,
    max_tokens=1024
)
```

```python
multi_lang_prompt = {
    
    "fr": """Vous êtes un assistant de mathématiques. Donnez le raisonnement et la réponse finale dans le format ci-dessous.
Question : Roger a 5 balles de tennis. Il achète 2 autres boîtes de balles de tennis. Chaque boîte contient 3 balles. Combien de balles a-t-il maintenant ?
Step-by-Step Answer : Roger started with 5 balls. 2 cans of 3 tennis balls each is 6 tennis balls. 5 + 6 = 11. The answer is 11.
Final Answer : 11
Question : {}
Step-by-Step Answer :  
Final Answer:""",
    
    "zh": """你是一个数学助手。请按以下格式给出推理过程和最终答案。
Question：Roger 有 5 个网球。他又买了 2 罐网球，每罐有 3 个网球。他现在有多少个网球？
Step-by-Step Answer：Roger started with 5 balls. 2 cans of 3 tennis balls each is 6 tennis balls. 5 + 6 = 11. The answer is 11.
Final Answer：11
Question：{}
Step-by-Step Answer：  
Final Answer：""",
    
    "es": """Eres un asistente de matemáticas. Da el razonamiento y la respuesta final en el siguiente formato.
Question: Roger tiene 5 pelotas de tenis. Compra 2 latas más de pelotas de tenis. Cada lata tiene 3 pelotas. ¿Cuántas pelotas tiene ahora?
Step-by-Step Answer: Roger started with 5 balls. 2 cans of 3 tennis balls each is 6 tennis balls. 5 + 6 = 11. The answer is 11.
Final Answer: 11
Question: {}
Step-by-Step Answer:  
Final Answer:""",
    
    "ko": """당신은 수학 조수입니다. 아래 형식으로 단계별 추론과 최종 답을 제공하세요.
Question: Roger는 5개의 테니스 공을 가지고 있습니다. 그는 테니스 공이 들어 있는 캔 2개를 더 샀습니다. 각 캔에는 3개의 테니스 공이 있습니다. 이제 그는 총 몇 개의 테니스 공을 가지고 있나요?
Step-by-Step Answer: Roger started with 5 balls. 2 cans of 3 tennis balls each is 6 tennis balls. 5 + 6 = 11. The answer is 11.
Final Answer: 11
Question: {}
Step-by-Step Answer:  
Final Answer:""",
    
    "hi": """आप एक गणित सहायक हैं। नीचे दिए गए प्रारूप में तर्क और अंतिम उत्तर दें।
Question: Roger के पास 5 टेनिस गेंदें हैं। वह 2 और कैन टेनिस गेंदों के खरीदता है। प्रत्येक कैन में 3 टेनिस गेंदें हैं। अब उसके पास कितनी टेनिस गेंदें हैं?
Step-by-Step Answer: Roger started with 5 balls. 2 cans of 3 tennis balls each is 6 tennis balls. 5 + 6 = 11. The answer is 11.
Final Answer: 11
Question: {}
Step-by-Step Answer:  
Final Answer:"""
}
```

```python
import re

def extract_final_answer(text: str):
    """
    Extract the last numeric value from text, including commas, decimals, and negatives.
    Example: "Answer is $70,000" -> "70,000"
    Returns the number as a string, or None if not found.
    """
    # Regex matches numbers with optional commas, decimals, and negatives
    numbers = re.findall(r'-?\d[\d,]*\.?\d*', text)
    
    if not numbers:
        return None
    return numbers[-1]  # return the last number

    
def compute_accuracy(predictions, gold_answers):
    correct = 0
    total = len(predictions)
    
    for pred_text, gold in zip(predictions, gold_answers):
        # extract last number from prediction
        pred_num = extract_final_answer(pred_text)
        # normalize commas for comparison
        if pred_num is not None:
            if pred_num.endswith('.'):
                pred_num = pred_num[:-1]
            pred_num_norm = pred_num.replace(",", "")
        else:
            pred_num_norm = None
        gold_norm = str(gold)
        
        if pred_num_norm == gold_norm:
            correct += 1
    accuracy = correct / total
    return accuracy
```

```python
from datasets import load_dataset
ds = load_dataset("juletxara/mgsm", "fr", split="test")

prompts = [multi_lang_prompt["fr"].format(ex["question"]) for ex in ds]
references = [ex["answer_number"] for ex in ds]

batch_size = 32
all_outputs = []

for i in range(0, len(prompts), batch_size):
    batch_prompts = prompts[i:i + batch_size]
    outputs = llm.generate(batch_prompts, sampling_params)
    for out in outputs:
        all_outputs.append(out.outputs[0].text.strip())
print(compute_accuracy(all_outputs, references))
```

```python
#hi
from datasets import load_dataset, Dataset
dataset = load_dataset("json", data_files="mgsm_multicolumn_multilang.jsonl", split="train")
ds2 = load_dataset("juletxara/mgsm", "en", split="test")
answers_list = ds2.to_list()
langs = ["hi"]
for lang in langs:
    question_field = f"question_{lang}"
    merged_items = []
    for i, item in enumerate(dataset):
        if question_field in item:
            answer_item = answers_list[i]
            answer = answer_item.get("answer_number")
            merged_items.append({
                "question": item[question_field],
                "answer_number": answer
            })
    
    # Convert merged list to a Hugging Face Dataset
    merged_ds = Dataset.from_list(merged_items)
prompts = [multi_lang_prompt["hi"].format(ex["question"]) for ex in merged_ds]
references = [ex["answer_number"] for ex in merged_ds]
batch_size = 32
all_outputs = []

for i in range(0, len(prompts), batch_size):
    batch_prompts = prompts[i:i + batch_size]
    outputs = llm.generate(batch_prompts, sampling_params)
    for out in outputs:
        all_outputs.append(out.outputs[0].text.strip())

print(compute_accuracy(all_outputs, references))
```
